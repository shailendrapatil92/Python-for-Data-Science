# How to use Pandas loc , iloc and ix

```python
import pandas as pd
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/shailendrapatil92/Data-Visualization/master/statewise_data.csv')
```

```python
data.head()
```

```python
data_2013 = data[data.Year == 2013]
```

```python
data_2013.reset_index(inplace=True)
```

```python
data_2013.drop(['index'],axis=1,inplace = True)
```

# data_2013 is our data for analysis

```python
data_2013.head()
```

## loc

```python
#When using label based indexing , both left and right values are inclusive
data_2013.loc[0:2,:]
```

```python
#As seen unlike range or numpy here 0,1,2 all 3 rows are considered i.e 0 and 2 both are inclusive
```

```python
#All rows but columns from Year to population
data_2013.loc[:,'Year':'TotalPopulation'].head()
```

```python
#We can mention the rows and columns in form of index
data_2013.loc[[0,1,4],:]
```

```python
data_2013.loc[[0,1,4],['Year','STATE','UnemployedPercent']]
```

```python
#Now Suppose say you want rows for state California, you might not know index
#We can make state as index and select california
data_2013.set_index('STATE').loc[['California'],:]
```

## iloc

```python
#position based
#Here indexing left value is inclusive right value exclusive
data_2013.head()
```

```python
data_2013.iloc[0:2,:]
```

```python
#As seen 0: 2 , 0 is inlcusive and 2 is exclusive 
#Apart from position and label based indexing this is other main difference
```

# ix

```python
#ix is used to do both index based and label based indexing
```

```python
data_2013.ix[0:2,0:2]
```

## Using loc for filtering

```python
data_2013.head()
```

```python
data_2013.loc[(data_2013.STATE == 'California') | (data_2013.STATE == 'Alaska'),:]
```
