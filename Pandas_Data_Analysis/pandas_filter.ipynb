# How do I filter rows of pandas dataframe by column value

```python
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
#Get dataframe where sepal_width greater then 3.5
data[data.sepal_width>3.5]
```

```python
#Multiple filter condition
#Get dataframe where sepal_width greater then 3.5 and species is virginica
data[(data.sepal_width>3.5) & (data.species == 'virginica')]
```
