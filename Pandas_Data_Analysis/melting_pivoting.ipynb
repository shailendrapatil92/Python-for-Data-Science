```python
import pandas as pd
```

```python
dic = {'semester':[1,2,3,4,5,6,7,8],'student_A':[3.9,4,3.3,3,3.7,3.8,3.9,3.95],'student_B':[3.7,4,3.6,3,3.6,3.9,3.2,3.1]}
```

```python
df = pd.DataFrame(dic)
```

```python
df.head()
```

```python
# the above dataset is semester grades for 2 students
#student_A and student_b can be melted down into single variable as student
```

```python
df_melt = pd.melt(frame=df,id_vars = 'semester',value_vars=['student_A','student_B'],var_name = 'Student',value_name='GPA')
#id_vars : column that should not be changed
#value_vars : columns that has to be melted
#var_name : what is the variable name that should be given to the new column
#value_name : column name for the values column
```

```python
df_melt
```

```python
#pivoting data
#Convert rows to columns : unique values into separate columns
df_pivot = df_melt.pivot(index='semester',columns = 'Student',values='GPA')
#index : Column to use to make new frame's index
#columns : colum name to make new dataframes columns
#values : values for new columns
```

```python
df_pivot.head()
```

```python
df_pivot.reset_index(inplace=True)
```

```python
df_pivot
```

```python
dict = {'semester':[2],'Student':['student_A'],'GPA':[2]}
new_df = pd.DataFrame(dict)
```

```python
# if there are duplicates in index then pivot throws error. 
# We need to use pivot table in such situation and handle duplicates
new_melted_df = pd.concat([df_melt,new_df],ignore_index=True)
```

```python
new_melted_df.head()
```

```python
df_pivot = new_melted_df.pivot(index='semester',columns = 'Student',values='GPA')
#As seen we get the error saying index contains duplicate values
```

```python
import numpy as np
df_pivot = new_melted_df.pivot_table(index='semester',columns = 'Student',values='GPA',aggfunc =np.mean)
#aggfunc : can specify how duplicate values has to be treated
```

```python
df_melt = pd.melt(frame=df,id_vars='year',value_vars=['male_0010','male_1020','female_0010','female_1020'],var_name='Sex',value_name='people_count')
```
